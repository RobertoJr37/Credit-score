Considere a mesma base de dados utilizada na tarefa 1 ```previsao_de_renda.csv```.

1) Com o que você viu em aula, faça uma análise das variáveis qualitativas (posse de imóvel e posse de veículo) para verificar se são boas preditoras de *renda*. Alguma delas parece discriminar *renda* de forma estatisticamente significante?

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats

# Carregar o arquivo CSV
df = pd.read_csv('previsao_de_renda.csv')

# Remover valores nulos em 'renda' para evitar problemas na análise
df = df.dropna(subset=['renda'])

# Boxplot para visualizar a distribuição da renda por posse de imóvel
plt.figure(figsize=(10, 6))
sns.boxplot(x='posse_de_imovel', y='renda', data=df)
plt.title('Renda por Posse de Imóvel')
plt.xlabel('Posse de Imóvel')
plt.ylabel('Renda')
plt.show()

# Boxplot para posse de veículo
plt.figure(figsize=(10, 6))
sns.boxplot(x='posse_de_veiculo', y='renda', data=df)
plt.title('Renda por Posse de Veículo')
plt.xlabel('Posse de Veículo')
plt.ylabel('Renda')
plt.show()

# Teste T para verificar diferença estatística entre as categorias de posse de imóvel
imovel_renda_sim = df[df['posse_de_imovel'] == 'sim']['renda']
imovel_renda_nao = df[df['posse_de_imovel'] == 'nao']['renda']
t_stat_imovel, p_val_imovel = stats.ttest_ind(imovel_renda_sim, imovel_renda_nao)

# Teste T para posse de veículo
veiculo_renda_sim = df[df['posse_de_veiculo'] == 'sim']['renda']
veiculo_renda_nao = df[df['posse_de_veiculo'] == 'nao']['renda']
t_stat_veiculo, p_val_veiculo = stats.ttest_ind(veiculo_renda_sim, veiculo_renda_nao)

# Mostrar resultados dos testes T
print(f"Teste T Posse de Imóvel: p-valor = {p_val_imovel}")
print(f"Teste T Posse de Veículo: p-valor = {p_val_veiculo}")
````

Nenhuma das variáveis qualitativas (posse de imóvel e posse de veículo) parece discriminar a renda de maneira estatisticamente significativa, com base na visualização dos boxplots. As distribuições são bastante semelhantes, sugerindo que esses fatores, isoladamente, não são bons preditores de renda.
